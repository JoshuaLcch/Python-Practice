```python
# -*- coding: utf-8 -*-
"""
Created on Wed Nov  7 18:46:16 2018

@author: joshua
"""

from PIL import Image
import numpy as np
from scipy import signal

def conv2(I,M):
    IH,IW = I.shape #高，寬
    MH,MW = M.shape
    out = np.zeros((IH-MH+1,IW-MW+1))
    for h in range(IH-MH+1):
        for w in range(IW-MW+1):
            for y in range(MH):
                for x in range(MW):
                    out[h,w] +=I[h+y,w+x]*M[y,x]
    return out

I = Image.open("/Users/joshua/Desktop/作品集整理/智慧型決策/drake.jpeg")
data = np.asarray(I)

data2 = data.copy()
R = data2[:,:,0]
G = data2[:,:,1]
B = data2[:,:,2]
#use mask 弄糊照片
Mask = np.ones((20,20))/400
#Mask = np.ones((1,60))/60 #快速移動!

#模糊照
R2 = signal.convolve2d(R, Mask, boundary = 'symm',mode = 'same')
G2 = signal.convolve2d(G, Mask, boundary = 'symm',mode = 'same')
B2 = signal.convolve2d(B, Mask, boundary = 'symm',mode = 'same')

#黑白照
R2 = (R+G+B)/3
G2 = (R+G+B)/3
B2 = (R+G+B)/3

data2[:,:,0] = R2
data2[:,:,1] = G2
data2[:,:,2] = B2
#R2 = conv2(R,Mask)
#G2 = conv2(G,Mask)
#B2 = conv2(B,Mask)
#data2[0:413,0:621,0] = R2
#data2[0:413,0:621,1] = G2
#data2[0:413,0:621,2] = B2

#data2[100:11,80:90,0]=255 #convert x,y axis' first bit (R) to 255
#data2[100:11,80:90,1]=0 #convert x,y axis' second bit (G) to 255
#data2[100:11,80:90,2]=0 #convert x,y axis' third bit (B) to 255

I2 = Image.fromarray(data2,'RGB')
I2.show()
```
